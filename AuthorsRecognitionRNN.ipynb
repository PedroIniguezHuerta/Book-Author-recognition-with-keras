## STEP 3: load dataset and do a quick cleanup of it

````
def removeShortLines(dataset, target):
  f = open(dataset,"r")
  lines = f.read()
  f.close()
  outputlines = ""
  llines = lines.split("\n")
  for line in llines:
    if line[-2:] == ";2":
      if np.random.randint(11) < 3:  
        continue
    line = re.sub('[^a-zA-z0-9;\s]','',line)
    line = line.lower().strip()
    if len(line) < 7:
      pass
    else:
      outputlines += line + "\n"
  f = open(target,"w")
  lines = f.write(outputlines)
  f.close()  
      
````

````
dataset_file = PATH + "/authors.csv"
target_file = "./authors.csv"
removeShortLines(dataset_file,target_file)
dataset = pd.read_csv(target_file,delimiter=";")

pd.set_option('display.max_colwidth', -1)
print(type(dataset['phrase'].str.lower()))
dataset['phrase'] = dataset['phrase'].str.lower()

dataset.head(20)

###################################################
# OPTIONAL: truncate dataset rows to make model light
###################################################
SAMPLES = 4000
dataset = dataset[:SAMPLES]
````

## STEP 11: Instance Do the whole prediction

In this part several steps are merged:
* Define the model
* Instane the model
* train the model
* plot results

The reason is because bellow provides and incremental approach where we can experience lot of combination of parameters. Progressivately the best combinations can be tried deeper and the worst combination discarded. Until having a unique one with the best results.

````
########################################################################
# Broad list of parameters supported.
########################################################################
#dUnits = [32,64,96,128,160,192,224,256]
#dDropout = [0.2, 0.3, 0.4, 0.7, 0.8, 0.9]
#dRecurrentDropout = [0.2, 0.3, 0.4, 0.7, 0.8, 0.9]
#dDense = [1, 2, 3, 4, 5]
#dActivation = ['relu', 'sigmoid','softmax','softplus','softsign','tanh', 'selu', 'elu', 'exponential']
#dOptimizer = ['adam', 'SGD','RMSprop', 'Adadelta', 'Adagrad', 'Adamax', 'Nadam','Ftrl']
#dLoss = ['sparse_categorical_crossentropy','binary_crossentropy','categorical_crossentropy',
#        'poisson','kl_divergence', 'mean_squared_error', 'mean_absolute_error', 'mean_absolute_percentage_error',
#         'mean_squared_logarithmic_error', 'cosine_similarity', 'huber', 'log_cosh', 'hinge', 'squared_hinge', 'categorical_hinge']

########################################################################
# More selective list of parameters to try after exploring parameters above
########################################################################
#dUnits = [64,128,256]
#dDropout = [0.4, 0.6, 0.8]
#dRecurrentDropout = [0.4, 0.8]
#dDense = [5]
#dActivation = ['sigmoid','softmax','softplus','softsign','tanh', 'selu', 'elu', 'exponential']
#dOptimizer = ['adam', 'RMSprop', 'Adadelta', 'Adagrad', 'Adamax', 'Nadam','Ftrl']
#dLoss = ['sparse_categorical_crossentropy','poisson','mean_absolute_percentage_error', 'categorical_hinge']


########################################################################
# Second selecton of parameters to try after exploring parameters above
########################################################################
dUnits = [64, 32, 128,256]
dDropout = [0.4]
dRecurrentDropout = [0.4]
dDense = [5]
dActivation = ['softplus','softmax', 'sigmoid']
dOptimizer = ['Nadam','Adam', 'Adagrad','RMSprop']
dLoss = [ 'sparse_categorical_crossentropy']

# This parameters can also be tried in several combinations.
BATCH_SIZE = 16
EPOCHS = 30
````

````
EMBEDDING_DIM = 100
unknown = len(tokenize.word_index)+1
models=[]

for pUnits in dUnits:
  for pDropout in dDropout:
    for dRecDrop in dRecurrentDropout:
      for pDense in dDense:
        for pActivation in dActivation:
          for pOptimizer in dOptimizer:
            for pLoss in dLoss:
              model = Sequential()
              model.add(Embedding(unknown, EMBEDDING_DIM, input_length=max_lenght))
              model.add(Bidirectional(LSTM(units=pUnits, dropout=pDropout, recurrent_dropout=dRecDrop)))
#              model.add(LSTM(units=pUnits, dropout=pDropout, recurrent_dropout=dRecDrop))
              model.add(Dense(pDense, activation=pActivation))
              model.compile(loss=pLoss, optimizer=pOptimizer, metrics=['accuracy'])
              models.append(model)

              print("Configuration parameters:")
              print("pUnits, pDropout, dRecDrop, pDense, pActivation, pOptimizer, pLoss")
              print(pUnits, pDropout, dRecDrop, pDense, pActivation, pOptimizer, pLoss)
              print(model.summary())

              try:
                history1=model.fit(X_train, y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=1, validation_split=0.2)
              except:
                continue

              print(history1.history['loss'])
              print(history1.history['accuracy'])
              print(history1.history['val_loss'])
              print(history1.history['val_accuracy'])


              epoch_count = range(1, len(history1.history['loss']) + 1)

              # Visualize learning curve. Here learning curve is not ideal. It should be much smoother as it decreases.
              #As mentioned before, altering different hyper parameters especially learning rate can have a positive impact
              #on accuracy and learning curve.
              plt.plot(epoch_count, history1.history['loss'], 'r-')
              plt.plot(epoch_count, history1.history['val_loss'], 'r--')
              plt.plot(epoch_count, history1.history['accuracy'], 'b-')
              plt.plot(epoch_count, history1.history['val_accuracy'], 'b--')
              plt.legend(['Training Loss','Validation Loss', 'Accuracy', 'Validation Accuracy'])
              plt.xlabel('Epoch')
              plt.ylabel('Loss')
              plt.show()
````
